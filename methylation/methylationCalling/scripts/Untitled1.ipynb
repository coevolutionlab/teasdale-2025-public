```r
library(ggplot2)
```

```r
df <- read.table("../output/at9852.test.nlrs.bedscore", 
                 header = F)
```

```r
colnames = c("chr","start","stop","geneName","gene","mCG")
```

```r
colnames(df) <- colnames
```

```r
head(df)
```

```r
ggplot(df, aes(x= gene, y = mCG)) +
geom_boxplot(outlier.fill = NA)+geom_jitter(width=0.2)
```

```r
df2 <- read.table("../output/at9852.test.nlrs.mCG.spo11.scores.txt", 
                  header = F,)
```

```r
colnames <- c("chr", "start", "stop", "gene", "mCG", "spo11")
```

```r
colnames(df2) <- colnames
```

```r
library(repr)
options(repr.plot.width=20, repr.plot.height=8)
ggplot(df2, aes(x= mCG, y= spo11, group = gene)) + geom_point(aes(col=gene), size = 5) + facet_wrap(~gene)
```
