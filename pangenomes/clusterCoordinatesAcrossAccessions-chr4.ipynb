```bash
echo "processing"


mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_gfa
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_allPathCoordinates
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_viz
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_pav
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_similarity
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_heaps
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_panacus
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_degrees
mkdir -p chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_depth
```

```bash
inputOdgi=chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/chr4.fasta.gz.856e650.e34d4cd.a91cbe1.smooth.final.og
```

```bash
odgi paths -L \
-i $inputOdgi
```

```bash
while read a b; 
do

    r=${a}
    outprefix=${b}.region

    ########################## view the pangenome graph of the region ######################
    odgi viz \
    -i $inputOdgi \
    -o chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_viz/${outprefix}.bm.png \
    -x 1000 \
    -bm \
    -t 20 \
    -P \
    -r ${r}
    ########################## Node Strandedness ############################################
    odgi viz \
    -i $inputOdgi \
    -o chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_viz/${outprefix}.z.png \
    -x 1000 \
    -z
    -t 20 \
    -P \
    -r ${r}
    
    ############################################ extract the region ############################################
    odgi extract \
    -i $inputOdgi \
    -r ${r} \
    --threads 24 -P \
    -d 200000 \
    -c 1  \
    -o chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.og 
    
    ########################### get the coordinates in all the accessions ########################################
    odgi paths -L \
    -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.og  \
    > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_allPathCoordinates/${outprefix}.allPathsCoordinates.txt
    
    
    ########################### for bandage plot #################################################################
    odgi view -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.og  \
    -g > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_gfa/${outprefix}.gfa
    
    ########################### Individual path-based PAV comparisons ############################################
    while read line;
    do
        r=${line}
        echo $r|cut -d ':' -f1,2 >tmp1
        echo $r|cut -d ':' -f3|sed "s/-/\t/g" >tmp2
        paste tmp1 tmp2 > \
        chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_allPathCoordinates/${outprefix}.${line}.bed && rm tmp1 && rm tmp2
        
        odgi pav \
        -i $inputOdgi \
        -t 24 \
        -b chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_allPathCoordinates/${outprefix}.${line}.bed > \
        chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_pav/${outprefix}.${line}.pav.txt
        
    done < chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_allPathCoordinates/${outprefix}.allPathsCoordinates.txt
    
    ########################### Node Jumps in Neighborhood ######################################################
    ## for each node how many jumps it has taken
    cat chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_gfa/${outprefix}.gfa \
    |grep "^L" \
    |awk '{print $1,$2,$4,$4-$2}'\
    |sed "s/ /\t/g" \
    > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_jumps/${outprefix}.nodeJumps.txt
    
    
    ########################### Node Degrees ############################################################
    odgi \
    degree \
    -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.og \
    -t 24 \
    -d \
    -r ${r} > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_degrees/${outprefix}.nodeDegrees.txt
    
    
    ########################### Node Depth ###############################################################
    odgi \
    depth \
    -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.og \
    -t 24 \
    -d \
    -r ${r} > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_depth/${outprefix}.nodeDepth.txt
    
    ########################### Distance between different Paths (accessions) for the region #############
    odgi \
    similarity \
    -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.og \
    -t 24 \
    -d > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_similarity/${outprefix}.distance.txt
    
    ###### Pangenome Graph Growth  Heaps' Law #########

    odgi sort \
    -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.og \
    -p Y \
    -O \
    -t 24 \
    -P \
    -o chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.optimized_sorted.og 
    
    for i in {1..18}
    do
    seq -w 1 ${i} > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_heaps/pathGroup${i}.groups.txt
    done
    
    
    for i in {1..18}
    do
    odgi paths -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.optimized_sorted.og \
    -L|sort -R|head -n ${i} > pathGroup${i}.txt 
    done
    
    for i in {1..18}
    do
    paste pathGroup${i}.txt pathGroup${i}.groups.txt > tmp
    mv tmp pathGroup${i}.txt
    done
    
    for i in {1..18}
    do
        odgi \
        heaps \
        -i chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_og/${outprefix}.optimized_sorted.og \
        -t 24 \
        -S \
        -d 2 \
        -n 100 \
        -p pathGroup${i}.txt \
        > chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/cluster_heaps/${outprefix}.pathGroup${i}.optimized_sorted.heaps.txt
    done
    
    
    ########################## Panacus ##############################
    parentDir=chr4_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended
    
    grep '^P' ${parentDir}/cluster_gfa/${outprefix}.gfa|cut -f2|sed "s/_.*//g" > ${parentDir}/cluster_panacus/${outprefix}.groupNames.txt
    grep '^P' ${parentDir}/cluster_gfa/${outprefix}.gfa|cut -f2> ${parentDir}/cluster_panacus/${outprefix}.pathNames.txt
    
    paste ${parentDir}/cluster_panacus/${outprefix}.pathNames.txt \
    ${parentDir}/cluster_panacus/${outprefix}.groupNames.txt \
    > ${parentDir}/cluster_panacus/${outprefix}.groups.txt
    
    sed -i "s/_1_chr4/#1#chr4/g" ${parentDir}/cluster_panacus/${outprefix}.pathNames.txt 
    sed -i "s/_1_chr4/#1#chr4/g" ${parentDir}/cluster_panacus/${outprefix}.groups.txt
    
    
    cat ${parentDir}/cluster_panacus/${outprefix}.pathNames.txt |cut -f1,3 -d ":" > \
    ${parentDir}/cluster_panacus/${outprefix}.newPathNames.txt
    
    paste ${parentDir}/cluster_panacus/${outprefix}.pathNames.txt \
    ${parentDir}/cluster_panacus/${outprefix}.newPathNames.txt > ${parentDir}/cluster_panacus/tmp
    
    while read a b ; do sed -i "s/${a}/${b}/" ${parentDir}/cluster_panacus/${outprefix}.pathNames.txt ; done < ${parentDir}/cluster_panacus/tmp
    while read a b ; do sed -i "s/${a}/${b}/" ${parentDir}/cluster_panacus/${outprefix}.groups.txt ; done < ${parentDir}/cluster_panacus/tmp
    cp ${parentDir}/cluster_gfa/${outprefix}.gfa ${parentDir}/cluster_panacus/${outprefix}.ed.gfa
    sed -i "s/_1_chr4/#1#chr4/g" ${parentDir}/cluster_panacus/${outprefix}.ed.gfa
    
    while read a b ; do sed -i "s/${a}/${b}/g" ${parentDir}/cluster_panacus/${outprefix}.ed.gfa ; done < ${parentDir}/cluster_panacus/tmp 
    
    panacus histgrowth -t4 -q 0,1,0.5,0.1 -g ${parentDir}/cluster_panacus/${outprefix}.groups.txt \
    -s ${parentDir}/cluster_panacus/${outprefix}.pathNames.txt \
    ${parentDir}/cluster_panacus/${outprefix}.ed.gfa > ${parentDir}/cluster_panacus/${outprefix}.histgrowth.txt
    
    panacus-visualize -e ${parentDir}/cluster_panacus/${outprefix}.histgrowth.txt > ${parentDir}/cluster_panacus/${outprefix}.histgrowth.node.pdf

done < chr4_cluster.neighborhoods.txt


```
