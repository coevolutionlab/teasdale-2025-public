```python
txtFile = f'{outvcfprefix}.txt'
separator="phased"

start = time.time()
df = pl.scan_csv(txtFile, sep="\t")

def convertDiploid(fileprefix, columnsToExclude=9, separator=str(separator)):
    outfile = f'{fileprefix}.diploid.txt'
    infile = f'{fileprefix}.txt'

    if separator == "phased":
        sepString = "|"
    elif separator == 'unphased':
        sepString = "/"
    else:
        print('wrong choice for separator (use <phased> or <unphased>)')

    df = pl.scan_csv(
        infile,
        sep="\t",
        null_values = {".":".|."},
        #ignore_errors = True,
    )
    
    (
        df
        .select
        (
            [
                pl.exclude(df.columns[columnsToExclude:]), # exclude first 9 columns
                pl.col(df.columns[columnsToExclude:]).apply(lambda x: f'{str(x)}{sepString}{str(x)}'), # make it pseudo-diploid
            ]
        )
        .collect()
        .write_csv(outfile, sep="\t",)
    )
    return 0


#executing function
convertDiploid(outvcfprefix,9,separator)
stop = time.time()
```
