## Odgi Bin

---
Per default, odgi bin writes the bins to stdout in a tab-delimited format: path.name, path.prefix, path.suffix, bin (bin identifier), mean.cov (mean coverage of the path in this bin), mean.inv (mean inversion rate of this path in this bin), mean.pos (mean nucleotide position of this path in this bin), first.nucl (first nucleotide position of this path in this bin), last.nucl (last nucleotide position of this path in this bin). These nucleotide ranges might span positions that are not present in the bin. Example: A range of 1-100 means that the first nucleotide has position 1 and the last has position 100, but nucleotide 45 could be located in another bin.


```bash
odgi bin \
-i chr3_cluster/out_p90_s5000/chr3.fasta.gz.92da240.e34d4cd.b17acf2.smooth.final.og \
-w 1000 \
-t 20 \
-P > test.odgi.bin.txt
```

#### Get the depth of the graph in 100 bp windows (using the 100 bp window bed)

**Above depth is mean depth in each 100 bp window.**

for each node within a window, the product between the node depth and its length (in bp) is calculated

Total sum is then divided by the length of the window (in bp) -> mean depth

```bash
odgi depth \
-i chr3_cluster/out_p90_s5000/chr3.fasta.gz.92da240.e34d4cd.b17acf2.smooth.final.og \
-b at9852_1_chr3.w100bps.bed | \
bedtools sort > at9852_1_chr3.w100bps.bed.depth
```

### Identify problematic regions and remove them
1. Low depth region (node depth of 0 and 1)
2. `-w 100:0:1:1` merges 100 bp into one region where node depth is 0 and 1 and TIPS are 1

```bash
### Identify low depth region
odgi depth -i chr3_cluster/out_p90_s5000/chr3.fasta.gz.92da240.e34d4cd.b17acf2.smooth.final.og \
-w 100:0:1:1 > chr3_cluster/out_p90_s5000/chr3.low_depth.bed
```

3. High depth region (node depth of 1000 and above)
4. region closer to 10000bp  merged into a single region
5. `-w 10000:1000:100000000`

```bash
odgi depth -i chr3_cluster/out_p90_s5000/chr3.fasta.gz.92da240.e34d4cd.b17acf2.smooth.final.og \
-w 10000:1000:100000000:0 > chr3_cluster/out_p90_s5000/chr3.high_depth.bed
```
