```bash
rm tmp*
while read chromosome gstart gstop
do
    echo ${chromosome} >> tmp3
    
    # take fifth leftmost node with depth.uniq=18 and length greater than 100 as left anchor for the neighborhood
    awk '($2>100)&&($4==18)' chr1_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/chr1.fasta.gz.856e650.e34d4cd.a91cbe1.smooth.final.og.nodeDepthLength.txt|awk -v "start=${gstart}" '($1<start)'|tail -n 5|head -n1|cut -f1 >> tmp1
    
    # take fifth rightmost node with depth.uniq=18 and length greater than 100 as right anchor for the neighborhood
    awk '($2>100)&&($4==18)' chr1_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/chr1.fasta.gz.856e650.e34d4cd.a91cbe1.smooth.final.og.nodeDepthLength.txt|awk -v "stop=${gstop}" '($1>stop)'|head -n 5|tail -n1|cut -f1 >> tmp2

done < neighborhoods/extra/chr1_eCompound.graph.bed

paste tmp3 tmp1 tmp2 > neighborhoods/extra/chr1_eNeighborhood.graph.bed
```

```bash
rm tmp*
while read chromosome nStart nStop
do
    echo ${chromosome} >> tmp3

    odgi position \
    -i chr1_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/chr1.fasta.gz.856e650.e34d4cd.a91cbe1.smooth.final.og \
    -t 24 \
    -g ${nStart} \
    -r ${chromosome} |tail -n +2|cut -d ',' -f4,5|sed "s/,/\t/g"|cut -f1 >> tmp1

    odgi position \
    -i chr1_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/chr1.fasta.gz.856e650.e34d4cd.a91cbe1.smooth.final.og \
    -t 24 \
    -g ${nStop} \
    -r ${chromosome} |tail -n +2|cut -d ',' -f4,5|sed "s/,/\t/g"|cut -f1 >> tmp2
done < neighborhoods/extra/chr1_eNeighborhood.graph.bed

paste tmp3 tmp1 tmp2 > neighborhoods/extra/chr1_eNeighborhood.coordinates.bed
```

```bash
inputOdgi=chr1_cluster.p95.s5000.n17.k47.G700-900-1100.at9852.pggbRecommended/chr1.fasta.gz.856e650.e34d4cd.a91cbe1.smooth.final.og
outdir=neighborhoods/chr1
count=0
while read a b c; 
do
    ((count+=1))   
    r_path=$(echo "${a}:${b}-${c}")
    r=${r_path}
    nh=$(echo "e_nh_${count}")
    outprefix=${nh}.region

    ########################## view the pangenome graph of the region ######################
    odgi viz \
    -i $inputOdgi \
    -o $outdir/${outprefix}.bm.png \
    -x 1000 \
    -bm \
    -t 20 \
    -P \
    -r ${r}
    ########################## Node Strandedness ############################################
    odgi viz \
    -i $inputOdgi \
    -o $outdir/${outprefix}.z.png \
    -x 1000 \
    -z \
    -t 20 \
    -P \
    -r ${r}
    
    ############################################ extract the region ############################################
    odgi extract \
    -i $inputOdgi \
    -r ${r} \
    --threads 24 -P \
    -d 100000 \
    -c 1  \
    -o $outdir/${outprefix}.og 
    
    ########################### get the coordinates in all the accessions ########################################
    odgi paths -L \
    -i $outdir/${outprefix}.og  \
    > $outdir/${outprefix}.allPathsCoordinates.txt
done < neighborhoods/extra/chr1_eNeighborhood.coordinates.bed
```

```bash
for i in {1..4}
do
sed -i "s/$/\tchr1_e_nh${i}/g" neighborhoods/chr1/e_nh_${i}.region.allPathsCoordinates.txt
done

for i in {1..4}
do 
sed -i "s/chr1_e_nh${i}/chr1_e_nh0${i}/g" neighborhoods/chr1/e_nh_${i}.region.allPathsCoordinates.txt
done


for i in {1..4}
do 
cut -d ":" -f1,3 neighborhoods/chr1/e_nh_${i}.region.allPathsCoordinates.txt |sed "s/:/\t/g"|sed "s/-/\t/g" > neighborhoods/chr1/e_nh_${i}.region.allPathsCoordinates.bed
done
```
